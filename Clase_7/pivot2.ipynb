```python
# Datos de ejemplo
data = {'Producto': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Mes': ['Enero', 'Enero', 'Febrero', 'Febrero', 'Enero', 'Febrero'],
        'Ventas': [100, 150, 200, 120, 110, 160]}
df = pd.DataFrame(data)

display(df)
```

```python
# Crear la tabla dinámica
tabla_pivot = pd.pivot_table(df, values='Ventas', index='Producto', columns='Mes', aggfunc='mean')

'''
Este crea una tabla donde las filas son los Producto y las columnas son el Mes.
La celda de cada producto y mes contiene el promedio (mean) de las ventas para
esa combinación.
'''
display(tabla_pivot)
```
